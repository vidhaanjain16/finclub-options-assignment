```python
import pandas as pd

def create_ensemble(svi_csv, imputer_csv, output_csv="submission_ensemble.csv"):
    #import previous submission files
    df_svi = pd.read_csv(svi_csv)
    df_imputer = pd.read_csv(imputer_csv)

    
    # Rename columns
    df_svi = df_svi.rename(columns={"value": "value_svi"})
    df_imputer = df_imputer.rename(columns={"value": "value_imputer"})
    
    # 2. Merge them together matching by the unique id
    df_ensemble = pd.merge(df_svi, df_imputer, on="id")
    # Calculate the Average
    df_ensemble["value"] = (df_ensemble["value_svi"] + df_ensemble["value_imputer"]) / 2

    
    # format for kaggle
    df_final = df_ensemble[["id", "value"]]
    
    #  Save the  file
    df_final.to_csv(output_csv, index=False)
    print(f" Ensemble submission saved to {output_csv}")

# run final
create_ensemble("submission_svi.csv", "submission_imputer.csv", "submission_ensemble.csv")
```
